# Imports

```python
from ultralytics import YOLOv10
import cv2
import matplotlib.pyplot as plt
import numpy as np
import math
```

# Model

```python
MODEL = "./weights/yolo/yolov10n.pt"

model = YOLOv10(MODEL)
```

# Image

```python
IMG = "./data/well_4_old.jpg"

img = cv2.imread(IMG)
```

# Bboxes

```python
def show_img(img, cmap=None):
    fig, ax = plt.subplots()

    ax.imshow(img, cmap=cmap)
    ax.axis("off")

    plt.show()
```

```python
step = 500
n_steps = math.ceil(img.shape[0] / step)

res_list = []

kernel = np.array([[-1, -1, -1], [-1, 9, -1], [-1, -1, -1]])

for i in range(6, n_steps):
    bottom = i * step
    top = (i + 1) * step

    img_slice = img[bottom : top]
    # show_img(img_slice)
    img_slice = cv2.morphologyEx(img_slice, cv2.MORPH_CLOSE, kernel=np.ones((3, 3), dtype=np.uint8), iterations=1)
    img_slice = cv2.blur(img_slice, (13, 13))
    img_slice = cv2.filter2D(img_slice, -1, kernel)

    # show_img(img_slice)

    # img_slice_res = cv2.resize(img_slice, (600, 224))

    # show_img(img_slice_res)
    # break

    res = model(img_slice)
    res_list.append(res)
```
